# Interpretasi Model
Menganalisis model terbaik: feature importance dan analisis overfitting (train vs test accuracy).

```python
import sys
sys.path.append('../src')

import joblib
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from utils import analyze_overfitting

best = joblib.load('../models/best_model.pkl')
```

## Feature Importance

```python
importances = best.named_steps['model'].feature_importances_ if 'model' in best.named_steps else best.named_steps['classifier'].feature_importances_
preprocessor_step = best.named_steps['pre'] if 'pre' in best.named_steps else best.named_steps['preprocessor']
feature_names = preprocessor_step.get_feature_names_out()

importance_df = pd.DataFrame({'Feature': feature_names, 'Importance': importances})
importance_df = importance_df.sort_values(by='Importance', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(data=importance_df.head(10), x='Importance', y='Feature')
plt.title('Top 10 Fitur Terpenting')
plt.show()
```

## Analisis Overfitting

```python
x_train = pd.read_csv('../data/processed/x_train.csv')
x_test = pd.read_csv('../data/processed/x_test.csv')
y_train = pd.read_csv('../data/processed/y_train.csv').iloc[:, 0]
y_test = pd.read_csv('../data/processed/y_test.csv').iloc[:, 0]

result = analyze_overfitting("Best Model", best, x_train, y_train, x_test, y_test)
result
```
